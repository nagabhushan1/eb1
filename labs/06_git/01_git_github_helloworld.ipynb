# üß≠ Git & GitHub Basics: Push a Simple `helloworld.py` from Local to GitHub

## üß© What is Git?
Git is a version control system that helps you track changes in your source code over time.

## üåê What is GitHub?
GitHub is a cloud-based platform that hosts Git repositories. It lets you store code online, collaborate, and use Git features like branching and pull requests.

## üå≥ What is the `main` Branch?
The main branch (formerly master) is the default branch in a Git repo. It represents the stable, production-ready version of your code.

### üñ•Ô∏è Step 1: Create a Folder for Your Project and change bash_profile to point to the workspace Folder
Open Command Prompt and run:
```bash
mkdir C:\workspace
mkdir C:\workspace\git
cd C:\workspace\git
```

```bash
pwd

cd /c/workspace/git

pwd
```

## üßæ Step 2: Create a Simple `helloworld.py` File
```bash
echo print("Hello, World!") > helloworld.py
```
Verify file content:
```bash
type helloworld.py
```

---
## ‚öôÔ∏è Step 3: Initialize Git Repository
Run in Git Bash:
```bash
git init
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
git config --list
```

---
## üì¶ Step 4: Add File and Commit
```bash
git add helloworld.py
git commit -m "Initial commit: Added helloworld.py"
git log --oneline
```

---
## ‚òÅÔ∏è Step 5: Create Repository on GitHub
Create a new repository on GitHub (without README) and copy its URL.

---
## üîó Step 6: Link Local Repo to GitHub
```bash
git remote add origin https://github.com/<your-username>/git-demo.git
git remote -v
```

---
## üöÄ Step 7: Push Code to GitHub
```bash
git branch -M main
git push -u origin main
```

---
## ‚úÖ Step 8: Verify on GitHub
Visit your repo and confirm that `helloworld.py` is uploaded.

# Orders_1NF
In **1NF**, we eliminate repeating groups and ensure atomic values.

### Problems in Unnormalized Table:
- `ProductList` had multiple products in one field (not atomic).

### Fix in 1NF:
- Each product gets its own row.
- Customer and Address still repeat (redundancy remains).

```python
CREATE TABLE Orders_1NF (
    OrderID NUMBER,
    CustomerName VARCHAR2(50),
    Address VARCHAR2(100),
    ProductName VARCHAR2(50),
    Quantity NUMBER,
    Price NUMBER
);
```
